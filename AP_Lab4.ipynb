```python
#python program to implement calculator using functions

def add(x, y):
    return x + y
def subtract(x, y):
    return x - y
def multiply(x, y):
    return x * y
def divide(x, y):
    if y != 0:
        return x / y
    else:
        return "Error! Division by zero."
num1 = 15
num2 = 5
print("Addition:", add(num1, num2))
print("Subtraction:", subtract(num1, num2))
print("Multiplication:", multiply(num1, num2))
print("Division:", divide(num1, num2))
```

```python
#python program to find factorial of a number using function
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
number = int(input("Enter a number to find its factorial: "))
result = factorial(number)
print("The factorial of", number, "is:", result)
```

```python
#python program to find area of circle, rectangle and triangle using function

def area_of_circle(radius):
    return 3.142 * radius * radius

def area_of_rectangle(length, width):
    return length * width
def area_of_triangle(base, height):
    return 0.5 * base * height
radius = 7
length = 10
width = 5
base = 8
height = 6
print("Area of Circle:", area_of_circle(radius))
print("Area of Rectangle:", area_of_rectangle(length, width))
print("Area of Triangle:", area_of_triangle(base, height))
```

```python
#simple interest using function
def simple_interest(principal, rate, time):
    interest = (principal * rate * time) / 100
    return interest
P = 1000  # Principal amount
R = 5     # Rate of interest
T = 3     # Time in years
SI = simple_interest(P, R, T)
print("Simple Interest is:", SI)        
```

```python
#convert celsius to fahrenheit using function
def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit      
celsius_temp = 25
fahrenheit_temp = celsius_to_fahrenheit(celsius_temp)
print(f"{celsius_temp}째C is equal to {fahrenheit_temp}째F")  
```

```python
#find largest of two numbers using function

def largest_of_two_numbers(a, b):
    if a > b:
        return a
    else:
        return b
    
largest = largest_of_two_numbers(45, 78)
print("The largest number is:", largest)    
```

```python
#convert kilometers to miles using function
def kilometers_to_miles(km):
    miles = km * 0.621371
    return miles
km_value = 10
miles_value = kilometers_to_miles(km_value)
print(f"{km_value} kilometers is equal to {miles_value} miles")
```
