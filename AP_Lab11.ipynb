```python
# python program to find array elements greaten than average

import numpy as np

arr = np.random.randint(1, 150, size=10)

print("Array elements:", arr)

total = 0

for i in range(len(arr)):
    total = total + arr[i]

avg = total / len(arr)

print("\n Array Sum = {0} and Average = {1}".format(total, avg))

print("\n Array elements greater than average:")

for i in range(len(arr)):
    if arr[i] > avg:
        print(arr[i], end=" ")
print()
        
```

```python
# A college wants to analyze students' marks in three subjects to find avarage, toppers, and pass percentage.

marks = np.array([
    [85, 92, 78],
    [88, 26, 95],
    [90, 91, 89],
    [70, 65, 80],
    [60, 75, 70]
])

# Average marks per student
average_marks = np.mean(marks, axis=1)
print("Average marks per student:", average_marks)

# Average marks per subject
average_subject_marks = np.mean(marks, axis=0)
print("Average marks per subject:", average_subject_marks)

# Student with highest total marks
toppers = np.max(marks, axis=0)
print("Toppers in each subject:", toppers)

# Pass percentage (pass mark is 40)
pass_count = np.sum(marks >= 40)
total_marks = marks.size
pass_percentage = (pass_count / total_marks) * 100
print("Pass percentage:", pass_percentage)

```

```python
# The factory produces 3 products in 4 shifts. Use numpy to calculate total and average production and identify the most productive shift.

production = np.array([
    [120, 150, 130],
    [140, 160, 135],
    [130, 155, 145],        
    [150, 170, 160]
])

# Total production
total_production = np.sum(production)
print("Total production:", total_production)

# Average production per shift
average_production = np.mean(production, axis=1)
print("Average production per shift:", average_production)

# Most productive shift
most_productive_shift = np.max(production, axis=1)
print("Most productive shift production:", most_productive_shift)
print("Most productive shift index:", np.argmax(np.sum(production, axis=1)))
```

```python
# Lab Exam

# A Teacher has two sets -- one for students who attended on monday and one for tuesday. Write a Python program to find who attended both, only MONDAY, AND ONLY Tuesday.

monday_attendance = {'Alice', 'Bob', 'Charlie', 'David'}
tuesday_attendance = {'Charlie', 'David', 'Eve', 'Frank'}

both_days = monday_attendance.intersection(tuesday_attendance)
only_monday = monday_attendance.difference(tuesday_attendance)
only_tuesday = tuesday_attendance.difference(monday_attendance)

print("Students who attended both days:", both_days)
print("Students who attended only Monday:", only_monday)
print("Students who attended only Tuesday:", only_tuesday)
```
