```python
# Pandas Series

import pandas as pd
a = [1, 7, 2]
myvar = pd.Series(a)
print(myvar)
```

```python
# Creating a series using Pandas Library

import pandas as pd
import numpy as np

# Creating empty series

ser = pd.Series()
print("Pandas Series:", ser)

# Simple array
data = np.array(['A', 'M', 'I', 'T', 'Y'])
ser = pd.Series(data)
print("Pandas Series:\n", ser)
```

```python
# Labels
print(myvar[2])
print(ser[3])
```

```python
# Create labels

myvar = pd.Series(a, index = ["x", "y", "z"])
print(myvar)
print(myvar['y'])
```

```python
# Key/value objects as series

calories = {'day1': 420, 'day2': 380, 'day3': 390}
myser = pd.Series(calories)
print(myser)
```

```python
myser = pd.Series(calories, index = ['day1', 'day2'])
print(myser)
```

```python
# Lab Exam

# Analyze daily stock prices of a company to find highest, lowest, and average closing price. using only numpy

import numpy as np

closing_prices = np.array([150.75, 152.30, 149.80, 153.40, 151.20, 154.10, 150.90])

highest_price = np.max(closing_prices)
lowest_price = np.min(closing_prices)
average_price = np.mean(closing_prices)

print("Highest Closing Price:", highest_price)
print("Lowest Closing Price:", lowest_price)
print("Average Closing Price:", average_price)
```

```python
# A weather department stores a week's temprature data(morning and evening). Analyze the tempreture difference and find the day with maximum variation.

import pandas as pd

data = {
    'Day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'Morning Temp': [15, 17, 16, 14, 18, 20, 19],
    'Evening Temp': [22, 21, 23, 20, 24, 25, 23]
}

df = pd.DataFrame(data)

df['Temp Difference'] = df['Evening Temp'] - df['Morning Temp']
max_variation_day = df.loc[df['Temp Difference'].idxmax()]


print("Day with maximum temperature variation:")
print(max_variation_day)
```

```python
# Lab Exam

# Analyze daily stock prices of a company to find highest, lowest, and average closing price.

import pandas as pd

data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'Closing Price': [150.0, 152.5, 148.0, 155.0, 149.5]
}

df = pd.DataFrame(data)

highest_price = df['Closing Price'].max()
lowest_price = df['Closing Price'].min()
average_price = df['Closing Price'].mean()

print(f"Highest Closing Price: {highest_price}")
print(f"Lowest Closing Price: {lowest_price}")
print(f"Average Closing Price: {average_price}")
```
