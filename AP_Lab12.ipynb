```python
# Write python program creating a database of employees using DataFrame method & dictionary.

import pandas as pd

data = {
    'Empnum': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age' : [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)
print(df)
```

```python
# reading data from csv file
df_csv = pd.read_csv('data.csv')
print(df_csv)
```

```python
print(df.head())
print(df.tail())
print(df.info())
```

```python
print(df.describe())
```

```python
ages = df['Age']
print("Ages of Employees:")
print(ages)
```

```python
name = df['Name']
print("Names of Employees:")
print(name)
```

```python
#display age geaterer than 18
age18 = df[df['Age'] > 18]
print("Employees with age greater than 18:")
print(age18)
```

```python
# adding new column
df['Salary'] = [50000, 60000, 70000]
print("DataFrame after adding Salary column:")
print(df)
```

```python
# adding a new column 'Designation'
df['Designation'] = ['Manager', 'Sr. Manager', 'Director']
print("DataFrame after adding Designation column:")
print(df)
```

```python
# Employee salary greater than 50000
high_salary = df[df['Salary'] > 50000]
print("Employees with salary greater than 50000:")
print(high_salary)
```

```python
# Print emnum and salary of employees
emp_salary = df[['Empnum', 'Salary']]
print("Employee Numbers and their Salaries:")
print(emp_salary)
```
