```python
# Write a python program to count even and odd numbers in a list using for loop

Numlist = []
Even_count = 0
Odd_count = 0

numbers = int(input("Enter the number of elements in the list: "))
for i in range(numbers):
    element = int(input(f"Enter element {i + 1}: "))
    Numlist.append(element)
    if element % 2 == 0:
        Even_count += 1
    else:
        Odd_count += 1

print("The list of numbers is:", Numlist)        
print("\nTotal even numbers:", Even_count)
print("Even numbers are:", [num for num in Numlist if num % 2 == 0])

print("\nTotal odd numbers:", Odd_count)
print("Odd numbers are:", [num for num in Numlist if num % 2 != 0])
```

```python
# python program to add and remove elements from a list

my_list = []
n = int(input("Enter the number of elements to add to the list: "))
for i in range(n):
    element = input(f"Enter element {i + 1}: ")
    my_list.append(element) 
print("List after adding elements:", my_list)
remove_element = input("Enter the element to remove from the list: ")
if remove_element in my_list:
    my_list.remove(remove_element)
    print("List after removing the element:", my_list)
else:
    print("Element not found in the list.") 
```

```python
#write a program to sort a list of integers in ascending and descending order using sort() method

int_list = []
n = int(input("Enter the number of integers to add to the list: "))
for i in range(n):
    element = int(input(f"Enter integer {i + 1}: "))
    int_list.append(element)
print("Original list:", int_list)
int_list.sort()
print("List sorted in ascending order:", int_list)
int_list.sort(reverse=True)
print("List sorted in descending order:", int_list)
```

```python
# pyton program to find sum of elements in a list

num_list = []
n = int(input("Enter the number of elements to add to the list: "))
for i in range(n):
    element = int(input(f"Enter element {i + 1}: "))
    num_list.append(element)
total_sum = sum(num_list)
print("The list of numbers is:", num_list)
print("Sum of all elements in the list:", total_sum)
```

```python
# Python program to add two lists example

list1 = []
list2 = []

n1 = int(input("Enter the number of elements for the first list: "))
for i in range(n1):
    element = input(f"Enter element {i + 1} for the first list: ")
    list1.append(element)

n2 = int(input("Enter the number of elements for the second list: "))
for i in range(n2):
    element = input(f"Enter element {i + 1} for the second list: ")
    list2.append(element)   
    
combined_list = list1 + list2
print("First list:", list1)
print("Second list:", list2)
print("Combined list:", combined_list)

```

```python
# python program to count positive and negative numbers in a list

num_list = []
positive_count = 0
negative_count = 0  
n = int(input("Enter the number of elements in the list: "))
for i in range(n):
    element = int(input(f"Enter element {i + 1}: "))
    num_list.append(element)
    if element >= 0:
        positive_count += 1
    else:
        negative_count += 1 
print("The list of numbers is:", num_list)        
print("\nTotal positive numbers:", positive_count)
print("Positive numbers are:", [num for num in num_list if num >= 0])   
print("\nTotal negative numbers:", negative_count)
print("Negative numbers are:", [num for num in num_list if num < 0])
```

```python
# Python program tom find largest element in a list

num_list = []
n = int(input("Enter the number of elements in the list: "))

for i in range(n):
    element = int(input(f"Enter element {i + 1}: "))
    num_list.append(element)

largest = max(num_list)

print("The list of numbers is:", num_list)
print("The largest element in the list is:", largest)
```
