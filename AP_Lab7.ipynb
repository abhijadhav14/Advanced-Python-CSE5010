```python
# Python program for Shopping Cart System: A user adds, removes and views items in a shopping cart.

cart = []

while True:
    print("\nShopping Cart Menu:")
    print("1. Add Item")
    print("2. Remove Item")
    print("3. View Cart")
    print("4. Exit")
    
    choice = input("Enter your choice (1-4): ")
    
    if choice == '1':
        item = input("Enter the item to add: ")
        cart.append(item)
        print(f"{item} has been added to your cart.")
        
    elif choice == '2':
        item = input("Enter the item to remove: ")
        if item in cart:
            cart.remove(item)
            print(f"{item} has been removed from your cart.")
        else:
            print(f"{item} is not in your cart.")
            
    elif choice == '3':
        if cart:
            print("Items in your cart:")
            for i, item in enumerate(cart, start=1):
                print(f"{i}. {item}")
        else:
            print("Your cart is empty.")
            
    elif choice == '4':
        print("Thank you for shopping!")
        break
        
    else:
        print("Invalid choice.")
```

```python
# Airline Passenger Check-in

passengers = []

while True:
    print("\nAirline Passenger Check-in Menu:")
    print("1. Add Passenger")
    print("2. Remove Passenger")
    print("3. View Passengers")
    print("4. Exit")
    
    choice = input("Enter your choice (1-4): ")
    
    if choice == '1':
        name = input("Enter the passenger's name to add: ")
        passengers.append(name)
        print(f"{name} has been added to the passenger list.")
        
    elif choice == '2':
        name = input("Enter the passenger's name to remove: ")
        if name in passengers:
            passengers.remove(name)
            print(f"{name} has been removed from the passenger list.")
        else:
            print(f"{name} is not in the passenger list.")
            
    elif choice == '3':
        if passengers:
            print("Current Passengers:")
            for i, name in enumerate(passengers, start=1):
                print(f"{i}. {name}")
        else:
            print("No passengers checked in.")
            
    elif choice == '4':
        print("Check-in process completed.")
        break
        
    else:
        print("Invalid choice.")
```

```python
#Python program to add an item to tupple

intTuple = (10, 20, 30, 40, 50)
print("Tuple items:", intTuple)

intTuple = intTuple + (70,)
print("Tuple items :", intTuple)

intTuple = intTuple + (80, 90)
print("Tuple items :", intTuple)

intTuple = intTuple[2:5] + (11, 22, 33, 44) + intTuple[7:]
print("Tuple items :", intTuple)
```

```python
# Python program to remove an item from tupple

numTuple = (9, 11, 22, 45, 67, 89, 15, 25, 19)
print("Tuple items :", numTuple)

numTuple = numTuple[:3] + numTuple[4:]
print("After removing 4th Tuple items :", numTuple)

numTuple = numTuple[:5] + numTuple[7:]
print("After removing 5th and 6th Tuple items :", numTuple)
```

```python
# Write a python Script to sum of tupple Even and Odd Numbers

numTuple = (10, 21, 32, 43, 54, 65, 76, 87, 98)
even_sum = 0
odd_sum = 0
for num in numTuple:
    if num % 2 == 0:
        even_sum += num
    else:
        odd_sum += num
print("Sum of Even Numbers :", even_sum)
print("Sum of Odd Numbers :", odd_sum)
```

```python
# To count positive and negative numbers in a tupple

numTuple = (10, -21, 32, -43, 54, -65, 76, 87, -98)
positive_count = 0
negative_count = 0
for num in numTuple:
    if num >= 0:
        positive_count += 1
    else:
        negative_count += 1 
print("Count of Positive Numbers :", positive_count)
print("Count of Negative Numbers :", negative_count)
```
