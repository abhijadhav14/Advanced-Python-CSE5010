```python
# WAP to create a dictionary.

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(thisdict)
```

```python
# Adding or modifying elements

thisdict["year"] = 2020
thisdict["color"] = "red"
print(thisdict)
```

```python
# Accessing elements
print(thisdict["brand"])
print(thisdict.get("model"))
```

```python
# Python Program to create 3 students details that has roll no and name in one disctionary

students = {
    1: "Bobby Hall",
    2: "John Cena",  
    3: "Emma Watson"
}   

for roll_no, name in students.items():
    print(f"Roll No: {roll_no}, Name: {name}")

students[4] = "Mike Tyson"
   
print("After adding a new student:")
for roll_no, name in students.items():
    print(f"Roll No: {roll_no}, Name: {name}")

# Python Disctionary get() Method
print(students.get(2))

# keys() method
print(students.keys())

```

```python
# Python program to check if a given key already exists in a dictionary using keys() method

key = int(input("Enter the key to check: "))
if key in students.keys():
    print(f"Key {key} exists in the dictionary.")
    print(f"key: {key}, value: {students[key]}")
else:
    print(f"Key {key} does not exist in the dictionary.")


# values() method

# numerical values
students1 = {1: 85, 2: 90, 3: 78}
print(students1.values())
# string values
print(students.values())

# items() method
print(students.items())
```

```python
# Sets
thisset = {"apple", "banana", "cherry"}
print(thisset)

# set of integers
thisset1 = {1, 2, 3, 4, 5}
print(thisset1)

# set of mixed datatypes
thisset2 = {1, "Hello", 3.4, True}
print(thisset2)
```

```python
# creating and modifying a set
thisset3 = set(("apple", "banana", "cherry"))
thisset3.add("orange")
print(thisset3)
```

```python
# Set Operations
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8} 
# Union
print(A | B)
# Intersection
print(A & B)
# Difference
print(A - B)
```
