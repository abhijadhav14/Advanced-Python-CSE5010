```python
#Python Program to find the factorial of a number using for loop

def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

number = int(input("Enter a number to find its factorial: "))
fact = factorial(number)
print(f"The factorial of {number} is {fact}")
```

```python
#Python Program to find sum and average of n natural numbers using while loop

def sum_and_average(n):
    total = 0
    count = 1
    while count <= n:
        total += count
        count += 1
    average = total / n
    return total, average

num = int(input("Enter a number to find sum and average of n natural numbers: "))
total_sum, avg = sum_and_average(num)
print(f"The sum of first {num} natural numbers is {total_sum}")
print(f"The average of first {num} natural numbers is {avg}")
```

```python
#Python Program to find sum and average of n natural numbers using for loop

def sum_and_average_for(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    average = total / n
    return total, average       

num_for = int(input("Enter a number to find sum and average of n natural numbers using for loop: "))
total_sum_for, avg_for = sum_and_average_for(num_for)
print(f"The sum of first {num_for} natural numbers is {total_sum_for}")
print(f"The average of first {num_for} natural numbers is {avg_for}")
```

```python
#nested for loop to print a pattern of 4 numbers

for i in range(1, 5):
    for j in range(i):
        print(i, end=' ')
    print()
```

```python
#nested for loop to print a pattern

n = int(input("Enter the number of rows for the pattern: "))
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print("*", end=" ")
    print()
```

```python
#nested for loop to print a pattern of stars after two spaces

n = int(input("Enter the number of rows for the star pattern: "))
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print("*", end="  ")
    print()

```

```python
#nested for loop to print a pattern of numbers in reverse order

for i in range(3, 0, -1):
    for j in range(i):
        print(i, end=' ')
    print()
```

```python
# parking lot space management

capacity = 5
cars = 0

while True:
    print("\n1. Park Car\n2. Remove Car\n3. Exit")
    choice = input("Enter your choice: ")

    if choice == '1':
        if cars < capacity:
            cars += 1
            print(f"Car parked. Cars inside: {cars}")
        else:
            print("Parking lot is full.")
    elif choice == '2':
        if cars > 0:
            cars -= 1
            print(f"Car removed. Cars inside: {cars}")
        else:
            print("Parking lot is empty.")
    elif choice == '3':
        print("Exiting the program.")
        break
    else:
        print("Invalid choice. Please try again.")
```

```python
# Write a python program to implement ATM cash withdrawal system as per the below requirements using loops and control statements:
# Ask user for PIN(3 chances)
# If PIN is correct -> allow withdrawal
# Stop after 3 wrong attempts 

correct_pin = "1234"
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    entered_pin = input("Enter your PIN: ")
    if entered_pin == correct_pin:
        print("PIN is correct. You can withdraw cash.")
        break
    else:
        attempts += 1
        print(f"Incorrect PIN. You have {max_attempts - attempts} attempts left.")
else:
    print("Maximum attempts reached. Your account is locked.")

    
```
